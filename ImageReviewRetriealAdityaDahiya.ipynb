```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
import string
import re
# import spacy

# Load dataset
df = pd.read_csv('A2_Data.csv')

# Load NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# # Load spaCy resources
# nlp = spacy.load("en_core_web_sm")

# Check for missing values in 'text_review' column
df = df.dropna(subset=['Review Text'])

# Lowercasing
df['processed_text'] = df['Review Text'].apply(lambda x: x.lower())

# Tokenization using NLTK
df['processed_text'] = df['processed_text'].apply(lambda x: word_tokenize(x))

# Removing punctuation
punctuation = string.punctuation
df['processed_text'] = df['processed_text'].apply(lambda x: [word for word in x if word not in punctuation])

# Stop word removal using NLTK
stop_words = set(stopwords.words('english'))
df['processed_text'] = df['processed_text'].apply(lambda x: [word for word in x if word not in stop_words])

# Stemming using NLTK Porter Stemmer
stemmer = PorterStemmer()
df['processed_text'] = df['processed_text'].apply(lambda x: [stemmer.stem(word) for word in x])

# # Lemmatization using spaCy
# def lemmatize_text(text):
#     doc = nlp(" ".join(text))
#     return [token.lemma_ for token in doc]

# df['processed_text'] = df['processed_text'].apply(lemmatize_text)

# df['processed_text'] = df['processed_text'].apply(lambda x: [word for word in x if word not in punctuation])

# Join tokens back into text
df['processed_text'] = df['processed_text'].apply(lambda x: ' '.join(x))

# Save or use the preprocessed data
df.to_csv('preprocessed_text_data.csv', index=False)
```

```python
import pandas as pd
import numpy as np
from collections import Counter
import math

# Load preprocessed data
df = pd.read_csv('preprocessed_text_data.csv')

# Calculate term frequency (TF)
def calculate_tf(tokens):
    tf = {}
    token_count = len(tokens)
    word_counts = Counter(tokens)
    for word, count in word_counts.items():
        tf[word] = count / token_count
    return tf

# Tokenize the preprocessed text
df['tokens'] = df['processed_text'].apply(lambda x: x.split())

# Calculate TF-IDF scores
def calculate_tfidf(tf, idf):
    tfidf = {}
    for word, tf_score in tf.items():
        tfidf[word] = tf_score * idf[word]
    return tfidf

# Calculate inverse document frequency (IDF)
def calculate_idf(documents):
    N = len(documents)
    idf = {}
    for document in documents:
        unique_words = set(document)
        for word in unique_words:
            if word in idf:
                idf[word] += 1
            else:
                idf[word] = 1
    for word, count in idf.items():
        idf[word] = math.log10(N / count)
    return idf

# Calculate TF-IDF scores for each document
tfidf_scores = []
documents = df['tokens'].tolist()
idf = calculate_idf(documents)
for tokens in documents:
    tf = calculate_tf(tokens)
    tfidf = calculate_tfidf(tf, idf)
    tfidf_scores.append(tfidf)

# Convert TF-IDF scores to DataFrame
tfidf_df = pd.DataFrame(tfidf_scores)

# Fill NaN values with 0 (for words not present in a document)
tfidf_df.fillna(0, inplace=True)

# Concatenate TF-IDF scores with the original DataFrame
result_df = pd.concat([df, tfidf_df], axis=1)

# Save or use the result
result_df.to_csv('tfidf_scores.csv', index=False)
```

```python
from PIL import Image, ImageEnhance
from PIL import Image, ImageFilter
from PIL import Image
import numpy as np

def change_brightness(img, brightness_factor):
    enhancer = ImageEnhance.Brightness(img)
    brightened_img = enhancer.enhance(brightness_factor)

    # Save the brightened image
    return brightened_img

def change_orientation(img):

    rotated_img = img.rotate(90)
    return rotated_img

def remove_grain(img, radius=2):
    denoised_img = img.filter(ImageFilter.MedianFilter())
    return denoised_img

def adjust_contrast(image, factor, output_image_path = "" ):
    enhancer = ImageEnhance.Contrast(image)
    adjusted_image = enhancer.enhance(factor)

    # Save the image
    return adjusted_image

def resize_image(img,size):
    resized_img = img.resize(size)

    return resized_img

def random_flips(img):
    # img = Image.open(input_image_path)

    if np.random.rand() < 0.5:
        img = img.transpose(Image.FLIP_LEFT_RIGHT)  # Flip horizontally
    # Randomly decide whether to flip vertically
    if np.random.rand() < 0.5:
        img = img.transpose(Image.FLIP_TOP_BOTTOM)
    return img
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def get_similar_image(image_embedding):

    product_number = 0
    similarity = {}
    while product_number < len(images_reviews_pairs):
        image_number = 0
        while image_number < len(images_reviews_pairs[product_number]["Images"]):
            image_now = (product_number, image_number)
            embedding_now = cosine_similarity(images_reviews_pairs[product_number]["Images"][image_number][1], image_embedding)[0][0]
            similarity[image_now] = embedding_now
            image_number += 1
        product_number += 1

    return similarity


def get_embedding(image):
    image = transform(image)     
    image = image.unsqueeze(0)   

    # Generate image embedding
    with torch.no_grad():
        embedding = resnet_model(image)

    # Flatten the embedding tensor
    embedding = torch.flatten(embedding, start_dim=1)

    # Convert embedding to numpy array
    embedding = embedding.numpy()

    return embedding
```
